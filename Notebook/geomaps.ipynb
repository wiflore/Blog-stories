```python
seattle_listing2 = seattle_listing.head(1000)
lati = seattle_listing2['latitude']
longi = seattle_listing2['longitude']
```

```python
sanfran_map = folium.Map(location=[lati[0], longi[0]], zoom_start=11)
incidents = folium.map.FeatureGroup()
 
# loop through the 100 crimes and add each to the incidents feature group
for lat, long, in zip(lati, longi):
    incidents.add_child(
        folium.CircleMarker(
            [lat, long],
            radius=2, # define how big you want the circle markers to be
            color='yellow',
            fill=True,
            fill_color='blue',
            fill_opacity=0.6
        )
    )
# add incidents to map
stationArr = seattle_listing2[['latitude', 'longitude']].values
sanfran_map.add_child(incidents)
sanfran_map.add_child(plugins.HeatMap(stationArr, radius=12))
```
